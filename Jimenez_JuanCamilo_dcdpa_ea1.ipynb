```python
import pandas as pd

import matplotlib.pyplot as plt

# Carga el archivo CSV en un DataFrame
df = pd.read_csv('/content/Reporte_Delito_Violencia_Intrafamiliar_Polic_a_Nacional.csv')
analisis_descriptivo = df.describe()
print("Análisis descriptivo:\n", analisis_descriptivo)

# Identificar los tipos de armas o medios utilizados
tipos_armas_medios = df['ARMAS MEDIOS'].unique()

# Mostrar los tipos de armas o medios utilizados
print("Tipos de armas o medios utilizados:")
for tipo in tipos_armas_medios:
    print("- ", tipo)


# Calcular la frecuencia de cada tipo de arma o medio utilizado
frecuencia_armas_medios = df['ARMAS MEDIOS'].value_counts()

# Graficar la frecuencia en un gráfico de torta
plt.figure(figsize=(8, 6))
plt.pie(frecuencia_armas_medios, labels=frecuencia_armas_medios.index, autopct='%1.1f%%', startangle=90)
plt.title('Frecuencia de tipos de armas o medios utilizados')
plt.axis('equal')
plt.show()

# Graficar la frecuencia en un gráfico de barras
plt.figure(figsize=(10, 6))
frecuencia_armas_medios.plot(kind='bar')
plt.xlabel('Tipo de arma o medio')
plt.ylabel('Frecuencia')
plt.title('Frecuencia de tipos de armas o medios utilizados')
plt.xticks(rotation=45)
plt.show()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calcular la suma de la columna "CANTIDAD" agrupada por "MUNICIPIO"
agregado = df.groupby('DEPARTAMENTO')['CANTIDAD'].sum().reset_index()

# Ordenar los datos por la columna "CANTIDAD" de manera descendente
agregado = agregado.sort_values('CANTIDAD', ascending=False)

# Crear el gráfico
plt.figure(figsize=(15, 5))
ax = sns.barplot(x='DEPARTAMENTO', y='CANTIDAD', data=agregado)

# Agregar los valores enteros a cada barra
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2, p.get_height()), ha='center', va='bottom', rotation='vertical')

plt.xticks(rotation=90)
# Agregar título al gráfico
plt.title('CANTIDAD DE HECHOS PRESENTADOS POR DEPARTAMENTOS')
plt.show()


print('\n')
# Calcular la suma de la columna "CANTIDAD" agrupada por "MUNICIPIO"
agregado = df.groupby('MUNICIPIO')['CANTIDAD'].sum().reset_index()

# Ordenar los datos por la columna "CANTIDAD" de manera descendente
agregado = agregado.sort_values('CANTIDAD', ascending=False)

# Seleccionar solo el top 15 de valores más altos
top_15 = agregado.head(15)

# Crear el gráfico
plt.figure(figsize=(15, 5))
ax = sns.barplot(x='MUNICIPIO', y='CANTIDAD', data=top_15)

# Agregar los valores enteros a cada barra
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2, p.get_height()), ha='center', va='bottom', rotation='vertical')

plt.xticks(rotation=90)
plt.title('CANTIDAD DE HECHOS PRESENTADOS POR MUNICIPIOS - TOP 15 ')
plt.show()


print('\n')

# Calcular la suma de la columna "CANTIDAD" agrupada por "GENERO"
agregado = df.groupby('GENERO')['CANTIDAD'].sum()

# Colores personalizados para cada categoría
colores = ['blue', 'pink']

# Crear el gráfico de torta
plt.figure(figsize=(6, 6))
plt.pie(agregado, labels=agregado.index, colors=colores, autopct='%1.1f%%')

# Agregar título al gráfico
plt.title('Distribución por Género')

plt.show()


print('\n')


# Calcular la suma de la columna "CANTIDAD" agrupada por "GRUPO ETARIO"
suma_cantidad = df.groupby('GRUPO ETARIO')['CANTIDAD'].sum()

# Ordenar los datos de mayor a menor
suma_cantidad = suma_cantidad.sort_values(ascending=True)

# Crear el gráfico de barras horizontales ordenado
plt.figure(figsize=(8, 6))
plt.barh(suma_cantidad.index, suma_cantidad, color=['red', 'orange', 'purple'])

# Agregar etiquetas para cada barra
for i, valor in enumerate(suma_cantidad):
    plt.text(valor, i, str(valor), ha='left', va='center')

# Agregar título y etiquetas de los ejes
plt.title('SUMA DE CANTIDAD POR GRUPO ETARIO')
plt.xlabel('CANTIDAD')
plt.ylabel('GRUPO ETARIO')

plt.show()
```
